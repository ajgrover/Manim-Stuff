```python
from manim import *
from scipy.optimize import fsolve
from manim.utils.color import Colors
from manim import RED, GREEN, BLUE
```

```python
%%manim -qm -v WARNING GraphingIntro

class GraphingIntro(Scene):
    def construct(self):
        # Define axes
        axes = Axes(
            x_range = (0, 1.2),
            y_range = (0, 1.8),
            x_length = 7,
            axis_config={"include_numbers": False, "include_ticks": False},
        )
        # Define graphs
        
        c=ValueTracker(0)
        
        cost_function = lambda x: x**2 + c.get_value()
        revenue_function = lambda x: x
        
        
        def optimum_finder(func1, func2, goToX=False):
            x_val = fsolve(lambda x: func1(x) - func2(x), 1)[0]
            if goToX:
                y_val = -.05 -.3*c.get_value()
            else:
                y_val = cost_function(x_val)
            return axes.c2p(x_val, y_val)

        def toRedraw():    
            lilLine = lambda x: x-.26+c.get_value()
            littleRev = axes.plot(lilLine, x_range=(.4, .6, 0.05), color=YELLOW)
            cost  = axes.plot(cost_function, x_range=(0, 1.1, 0.05))
            cost_label = axes.get_graph_label(cost, label=r"\text{Total Private}\\ \text{Cost}", x_val=1.1, direction=UP, buff=0).scale(.4)
            revenue = axes.plot(revenue_function, x_range=(0, 1.1, 0.05))
            revenue_label = axes.get_graph_label(revenue, label=r"\text{Total Private}\\ \text{Revenue}", x_val=1.1, direction=RIGHT, buff=-.5).scale(.4)


            optimal_lil = axes.get_vertical_line(axes.c2p(0.5, 0.25+c.get_value()), color=YELLOW)
            optimum_lil_label = Tex(r"$I^*_{P}$", font_size=22, color=YELLOW).move_to(axes.c2p(0.5, -.05))
        
            return VGroup(cost, revenue, cost_label, revenue_label, littleRev, optimal_lil, optimum_lil_label)
        
        funcs = always_redraw(toRedraw)
        
        def reoptimize():
            optimal_point = optimum_finder(cost_function, revenue_function)
            optimal = axes.get_vertical_line(optimal_point, color=BLUE)
            dot = Dot(point=optimal_point, color=BLUE)
            optimal_point_label = optimum_finder(cost_function, revenue_function, True)
            optimum_label = Tex(r"$I_{OA}$", font_size=22, color=BLUE).move_to(optimal_point_label)
            return VGroup(optimal, dot, optimum_label)
        
        plot_opt = always_redraw(reoptimize)

        axes_labels = axes.get_axis_labels(Tex(r"Resources Required\\ to Deforest").scale(0.6), Tex("Total Monetary Value").scale(0.6))
        
        
        k = ValueTracker(0)
        
        profit_label = Tex(r"profit", font_size=16, color=ORANGE).move_to(axes.c2p(.55, .41))
        tax_label = Tex(r"tax", font_size=16, color=ORANGE).move_to(axes.c2p(-.04, .14))
        
        def taxes():
            dot = Dot(point=axes.c2p(.5-k.get_value(), cost_function(.5)-.5*k.get_value()), color=ORANGE)
            dot2 = Dot(point=axes.c2p(.5-k.get_value(), revenue_function(.5)-.5*k.get_value()), color=ORANGE)
            line = Line(dot.get_center(), dot2.get_center()).set_color(ORANGE)
            return VGroup(line)
        
        tax_group = always_redraw(taxes)
           
        self.add(
            axes,
            axes_labels,
            funcs,
            plot_opt
        )
        
        self.wait(1)
        self.play(FadeIn(tax_group, profit_label))
        self.wait(1.5)
        self.play(k.animate.set_value(.5),
                 FadeTransform(profit_label, tax_label)
                 )
        self.wait(1)
        self.play(c.animate(run_time=2).set_value(0.25))
        self.wait(1.5)
```

```python
%%manim -qm -v WARNING GraphingIntro

class GraphingIntro(Scene):
    def construct(self):
        # Define axes
        axes = Axes(
            x_range = (0, 1.2),
            y_range = (0, 2.2),
            x_length = 7,
            axis_config={"include_numbers": False, "include_ticks": False},
        )
        # Define graphs
        
        c=ValueTracker(0)
        s=ValueTracker(1)
        
        cost_function = lambda x: s.get_value()*x**2 + c.get_value()
        revenue_function = lambda x: x
        damage_func = lambda x: (1/2)*x**2
        social_func = lambda x: cost_function(x) + damage_func(x)
        soclil = lambda x: x-social_func(1/3)-.01 #rev func adjusted
        
        
        def optimum_finder(func1, func2, goToX=False):
            x_val = fsolve(lambda x: func1(x) - func2(x), 1)[0]
            if goToX:
                y_val = -.05 -.3*c.get_value()
            else:
                y_val = cost_function(x_val)
            return axes.c2p(x_val, y_val)

        def toRedraw():
            cost  = axes.plot(cost_function, x_range=(0, 1.1, 0.05))
            cost_label = axes.get_graph_label(cost, label=r"\text{Total Private}\\ \text{Cost}", x_val=1.1, direction=UP, buff=0).scale(.4)
            revenue = axes.plot(revenue_function, x_range=(0, 1.1, 0.05))
            revenue_label = axes.get_graph_label(revenue, label=r"\text{Total Private}\\ \text{Revenue}", x_val=1.1, direction=RIGHT, buff=-.5).scale(.4)
            return VGroup(cost, revenue, cost_label, revenue_label)
        
        funcs = always_redraw(toRedraw)
        
        def yellowBoys():
            lilLine = lambda x: x-.26+c.get_value()
            littleRev = axes.plot(lilLine, x_range=(.45, .55, 0.05), color=YELLOW)
            optimal_lil = axes.get_vertical_line(axes.c2p(0.5, 0.25+c.get_value()), color=YELLOW)
            optimum_lil_label = Tex(r"$I^*_{P}$", font_size=22, color=YELLOW).move_to(axes.c2p(0.5, -.05))
            return VGroup(optimal_lil, optimum_lil_label, littleRev)
        
        yellowFuncs = always_redraw(yellowBoys)
        
        def reoptimize():
            optimal_point = optimum_finder(cost_function, revenue_function)
            optimal = axes.get_vertical_line(optimal_point, color=BLUE)
            dot = Dot(point=optimal_point, color=BLUE)
            optimal_point_label = optimum_finder(cost_function, revenue_function, True)
            optimum_label = Tex(r"$I_{OA}$", font_size=22, color=BLUE).move_to(optimal_point_label)
            return VGroup(optimal, dot, optimum_label)
        
        plot_opt = always_redraw(reoptimize)

        axes_labels = axes.get_axis_labels(Tex(r"Resources Required\\ to Deforest").scale(0.6), Tex("Total Monetary Value").scale(0.6))
        
        
        k = ValueTracker(0)
        
        profit_label = Tex(r"profit", font_size=16, color=ORANGE).move_to(axes.c2p(.38, .29))
        tax_label = Tex(r"tax", font_size=16, color=ORANGE).move_to(axes.c2p(-.04, .11))
        
        def taxes():
            dot = Dot(point=axes.c2p((1/3)-k.get_value(), social_func(1/3)-.5*k.get_value()), color=ORANGE)
            dot2 = Dot(point=axes.c2p((1/3)-k.get_value(), revenue_function(1/3)-.5*k.get_value()), color=ORANGE)
            line = Line(dot.get_center(), dot2.get_center()).set_color(ORANGE)
            return VGroup(line)
        
        tax_group = always_redraw(taxes)
        
        
        # FIGURE 2 ADDED LINES AND STUFF:
        
        def drawFig2():
            lil = axes.plot(soclil, x_range=((1/3)-.1, (1/3)+.1, 0.05), color=GREEN)
            dmg = axes.plot(damage_func, x_range=(0, 1.1, 0.05), color=Colors.gold_c.value)
            dmg_label = axes.get_graph_label(dmg, label=r"\text{Total Damage}\\ \text{Cost}", x_val=.95, direction=UR, buff=0).scale(.4)
            soc = axes.plot(social_func, x_range=(0, 1.1, 0.05), color=Colors.red_c.value)
            soc_label = axes.get_graph_label(soc, label=r"\text{Total Social}\\ \text{Cost}", x_val=1.2, direction=UP, buff=0).scale(.4)
            
            optsoc = axes.get_vertical_line(axes.c2p((1/3), social_func(1/3)), color=GREEN)
            optsoc_label = Tex(r"$I^*_{S}$", font_size=22, color=GREEN).move_to(axes.c2p((1/3), -.05))
            
            return VGroup(lil, dmg, dmg_label, soc, soc_label, optsoc, optsoc_label)
        
        
        fig2 = drawFig2()
        
           
        self.add(
            axes,
            axes_labels,
            funcs,
            plot_opt,
            yellowFuncs
        )
        self.play(FadeIn(fig2, run_time=2))
        self.wait(2)
        self.wait(1)
        self.play(FadeIn(tax_group, profit_label))
        self.wait(1.5)
        self.play(k.animate.set_value(1/3),
                 FadeTransform(profit_label, tax_label)
                 )
        self.wait(1)
        self.play(c.animate(run_time=2).set_value(revenue_function(1/3)-social_func(1/3)))
        self.wait(1.5)
        self.play(FadeOut(yellowFuncs),
                  s.animate(run_time=2).set_value(1.5)
                 )
        self.wait(2)
        
        
        
```
